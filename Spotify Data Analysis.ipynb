```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
# Checking for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# Finding the least popular songs
sorted_df = df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
# top 10 songs
most_popular_songs = df_tracks[df_tracks['popularity']>90]
most_popular_songs.sort_values('popularity',ascending=False).iloc[:11]
```

```python
df_tracks.set_index('release_date',inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Converting duration_ms to seconds
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
# droping the additional duration_ms column
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
df_tracks.head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='crest',linewidths=1,
                      linecolor='Black')
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# Creating a regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'loudness',x = 'energy', color = 'c').set(title = 'Loudness Vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'popularity',x = 'acousticness', color = 'c').set(title = 'Popularity Vs Acousticness Correlatio')
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
# Visualizing the songs 
sns.displot(years, discrete=True, aspect = 2, height=6,kind = 'hist').set(title = "Number of songs per year")
```

```python
total_duration = df_tracks['duration']
fig_dims = (18,7)

fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title = 'Year Vs Duration')
plt.xticks(rotation = 90)
```

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y= 'genre', x='duration_ms',data = df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10,7))
popular_songs = df_genre.sort_values('popularity',ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = popular_songs).set(title = 'Top 5 Genres based on Popularity')
```
